```python
# Lucky Choudhary

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv("../data/processed/final_walmart.csv")
```

```python
print(df.shape)
print(df.columns)
df.head()
```

```python
sns.set(style="whitegrid")
```

```python
os.makedirs("../images", exist_ok=True)
```

```python
plt.figure()
sns.histplot(df['Weekly_Sales'], bins=50)
plt.title("Sales Distribution")

plt.savefig("../images/sales_distribution.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['Weekly_Sales'])
plt.title("Sales Outliers")

plt.savefig("../images/sales_boxplot.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure()
df.groupby('Month')['Weekly_Sales'].mean().plot()
plt.title("Monthly Sales Trend")

plt.savefig("../images/monthly_trend.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure()
sns.barplot(x='Type', y='Weekly_Sales', data=df)
plt.title("Sales by Store Type")

plt.savefig("../images/store_type.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure()
sns.barplot(x='IsHoliday', y='Weekly_Sales', data=df)
plt.title("Holiday vs Sales")

plt.savefig("../images/holiday_sales.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure()
df.groupby('Store')['Weekly_Sales'].sum().head(10).plot(kind='bar')
plt.title("Top Stores by Sales")

plt.savefig("../images/top_stores.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='Size', y='Weekly_Sales', data=df)
plt.title("Store Size vs Sales")

plt.savefig("../images/store_size_vs_sales.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure()
df.groupby('Dept')['Weekly_Sales'].mean().head(10).plot(kind='bar')
plt.title("Top Departments by Sales")

plt.savefig("../images/dept_sales.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure()
df.groupby('Dept')['Weekly_Sales'].mean().head(10).plot(kind='bar')
plt.title("Top Departments by Sales")

plt.savefig("../images/dept_sales.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure()
sns.histplot(df['Total_MarkDown'], bins=50)
plt.title("Discount Distribution")

plt.savefig("../images/discount_distribution.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='Discount_Ratio', y='Weekly_Sales', data=df)
plt.title("Discount Ratio vs Sales")

plt.savefig("../images/discount_ratio.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure()
df.groupby('Quarter')['Weekly_Sales'].sum().plot(kind='bar')
plt.title("Quarterly Sales")

plt.savefig("../images/quarterly_sales.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure()
df.groupby('Quarter')['Weekly_Sales'].sum().plot(kind='bar')
plt.title("Quarterly Sales")

plt.savefig("../images/quarterly_sales.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure()
sns.histplot(df['Performance_Score'], bins=50)
plt.title("Performance Score Distribution")

plt.savefig("../images/performance_score.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='Sales_per_Size', y='Weekly_Sales', data=df)
plt.title("Efficiency vs Sales")

plt.savefig("../images/efficiency.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='Temperature', y='Weekly_Sales', data=df)
plt.title("Temperature vs Sales")

plt.savefig("../images/temp_vs_sales.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='Fuel_Price', y='Weekly_Sales', data=df)
plt.title("Fuel Price vs Sales")

plt.savefig("../images/fuel_vs_sales.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='Unemployment', y='Weekly_Sales', data=df)
plt.title("Unemployment vs Sales")

plt.savefig("../images/unemployment.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(), cmap='coolwarm')
plt.title("Correlation Heatmap")

plt.savefig("../images/heatmap.png", bbox_inches='tight')
plt.show()
```

```python
plt.figure()
sns.countplot(x='Sales_Category', data=df)
plt.title("Sales Category Distribution")

plt.savefig("../images/sales_category.png", bbox_inches='tight')
plt.show()
```
