```python
# Lucky Choudhary

import pandas as pd

# Load all datasets from 'data/raw' folder 
train = pd.read_csv("../data/raw/train.csv")
features = pd.read_csv("../data/raw/features.csv")
stores = pd.read_csv("../data/raw/stores.csv")

# Preview(shows first five rows from all the loaded files from above)
print(train.head())
print(features.head())
print(stores.head())
```

```python
# Preview helps to understand the structure of raw data
print(train.info())
print(train.describe())

print(features.info())
print(stores.info())
```

```python
# Check missing values(in train file)
print(train.isnull().sum())

# Check duplicates(in train file)
print(train.duplicated().sum())

# Check missing values(in features file)
print(features.isnull().sum())

# Check duplicates(in features file)
print(features.duplicated().sum())

# Check missing values(in stores file)
print(stores.isnull().sum())

# Check duplicates(in stores file)
print(stores.duplicated().sum())
```
