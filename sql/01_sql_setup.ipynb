```python
# Lucky Choudhary
# Setting up SQL DataBase, so it's more effecient and productive to analyze data int future queries 
import sqlite3
import pandas as pd
import os

# Load final cleaned data
df = pd.read_csv("../data/processed/final_walmart.csv")

# Create database folder
os.makedirs("../database", exist_ok=True)

# Create database
db_path = "../database/walmart.db"
conn = sqlite3.connect(db_path)

# Save table
df.to_sql("sales", conn, if_exists="replace", index=False)

conn.close()

print("Database Ready!")
```
