## Top 5 Stores by Total Sales

This query identifies the top-performing stores based on total sales.

Business Goal:
To understand which stores generate the highest revenue and contribute most to overall performance.

```python
query = """
SELECT Store, SUM(Weekly_Sales) AS total_sales
FROM sales
GROUP BY Store
ORDER BY total_sales DESC
LIMIT 5;
"""
pd.read_sql(query, conn)
```

## Top 5 Departments by Sales

This query highlights the departments with the highest total sales.

Business Goal:
To identify which product categories drive the most revenue.

```python
query = """
SELECT Dept, SUM(Weekly_Sales) AS total_sales
FROM sales
GROUP BY Dept
ORDER BY total_sales DESC
LIMIT 5;
"""
pd.read_sql(query, conn)
```

## Holiday vs Non-Holiday Sales Analysis

This query compares average sales during holiday and non-holiday periods.

Business Goal:
To analyze the effect of holidays on customer purchasing behavior.

```python
query = """
SELECT IsHoliday, AVG(Weekly_Sales) AS avg_sales
FROM sales
GROUP BY IsHoliday;
"""
pd.read_sql(query, conn)
```

## Monthly Sales Trend

This query analyzes average sales across different months.

Business Goal:
To identify seasonal patterns and peak sales months.

```python
query = """
SELECT Month, AVG(Weekly_Sales) AS avg_sales
FROM sales
GROUP BY Month
ORDER BY Month;
"""
pd.read_sql(query, conn)
```

## Sales by Store Type

This query compares average sales across different store types.

Business Goal:
To evaluate how store classification impacts performance.

```python
query = """
SELECT Type, AVG(Weekly_Sales) AS avg_sales
FROM sales
GROUP BY Type;
"""
pd.read_sql(query, conn)
```

## Stores with Highest Discount Dependency

This query identifies stores that rely heavily on discounts.

Business Goal:
To understand how much stores depend on promotions to drive sales.

```python
query = """
SELECT Store, AVG(Discount_Ratio) AS avg_discount
FROM sales
GROUP BY Store
ORDER BY avg_discount DESC
LIMIT 5;
"""
pd.read_sql(query, conn)
```

## Best Performing Stores (Adjusted Performance)

This query ranks stores based on performance score, which accounts for discount impact.

Business Goal:
To identify stores that perform well without heavy reliance on discounts.

```python
query = """
SELECT Store, AVG(Performance_Score) AS performance
FROM sales
GROUP BY Store
ORDER BY performance DESC
LIMIT 5;
"""
pd.read_sql(query, conn)
```

## Quarterly Sales Analysis

This query shows total sales across different quarters.

Business Goal:
To analyze seasonal trends and business cycles.

```python
query = """
SELECT Quarter, SUM(Weekly_Sales) AS total_sales
FROM sales
GROUP BY Quarter
ORDER BY Quarter;
"""
pd.read_sql(query, conn)
```

## Sales vs Store Efficiency

This query compares total sales with sales efficiency (sales per store size).

Business Goal:
To identify whether high sales come from efficiency or just large store size.

```python
query = """
SELECT Store,
       AVG(Weekly_Sales) AS sales,
       AVG(Sales_per_Size) AS efficiency
FROM sales
GROUP BY Store
ORDER BY sales DESC
LIMIT 10;
"""
pd.read_sql(query, conn)
```
